```python
# apple, lobby, daddy, right, panic, light, heavy, heart, dvice, dhict
import numpy as np
data = np.array(
    [
        [a,p,p,l,e],
        [l,o,b,b,y],
        [d,a,d,d,y],
        [r,i,g,t,t],
        [p,a,n,i,c],
        [l,i,g,h,t],
        [h,e,a,v,y],
        [h,e,a,r,t],
        [d,v,i,c,e],
        [d,h,i,c,t]
    ]
)
```

```python
# 시계열 데이터 : 데이터가 순서를 가지고 있다.(보통 3차원)
# sample, timestep, feature
# sample(10) : 가지고 있는 데이터의 갯수
# timestep(5) : 하나의 데이터가 가지고 있는 순서
# feature(26) : 하나의 timestep이 표현할 수 있는 갯수
data.shape
```

```python
# 앞 4글자는 문제 데이터
# 뒤 1글자는 정답 데이터
X = data[: , :-1]
y = data[: , -1]
X.shape, y.shape
```

```python
# 딥러닝 모델
# MLP, CNN, YOLO > 여러개의 데이터를 학습할 수 있다
#                > 순서에 따라 의미가 달라지는 데이터는 학습은 됨, 결과가 안나옴
#                > 순서를 상관하지 않고 학습

# 시계열 딥러닝 모델
# RNN, LSTM > 여러 데이터를 학습할 수 있음
#           > 순서에 따라 의미가 달라지는 데이터는 학습은 됨, 결과 잘 나옴
#           > 순서를 기억한채 학습
```

```python
# 모델링
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN

model = Sequential()

# input_shape = 입력 데이터의 크기(갯수는 상관하지 않는다, 데이터중 1개를 의미)
# activation = 'tanh' : 학습을 오래 진행하면 전에 있던 데이터를 잊는다, 임팩트 있는 값들은 기억하지만 아닌 것들은 X
model.add(SimpleRNN(units=30, input_shape = (4,26), activation = 'tanh'))

# units = 26 : 알파벳의 갯수 > 다중분류 (정답으로 말해줄 수 있는 데이터가 a-z까지 있기 때문에 26개)
model.add(Dense(units = 26, activation = 'softmax'))

# 학습방법 설정
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

# 학습
model.fit(X,y,epochs = 300)
```

```python
# apple, lobby, daddy, right, panic, light, heavy, heart, dvice, dhict
model.predict([[l,o,b,b]])
```
